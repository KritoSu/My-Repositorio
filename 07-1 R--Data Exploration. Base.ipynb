**Haga un histograma simple de la variable `Sepal.Length` de la tabla (data.frame) `iris` usando el sistema base.**

```r
str(iris)
```

**Personalice el histograma anterior.**

```r
hist(iris$Sepal.Length,                   # la variable
     col=rainbow(20),                     # color del histograma
     xlab = "TItulo eje X",               # titulo del eje X
     ylab = "Titulo eje Y",               #¬†titulo del eje Y
     main = "Titulo principal",           # titulo del grafico
     xlim = c(4, 9),                      # limites eje X
     ylim = c(0, 20),                     # limite eje Y
     breaks = 15)                         # n√∫mero de bins

rug(iris$Sepal.Length)                    # datos en el eje X
abline(h = 10, lwd = 2)                   # linea horizontal y = 10
abline(v = 4.9, lwd = 3, col='red')       # linea vertical
grid()                                    #¬†grid (lineas punteadas)
text(7,19, c('Linea 1\n', 'Linea 2'))
arrows(7, 19, 5, 15, col = 'black', lwd = 2)
```

**Haga un boxplot de la variable Sepal.Length.**

**Haga un scatterplot de Sepal.Length vs Petal.Length y agregue una linea de regresi√≥n.**

```r
plot(Sepal.Length ~ Petal.Length, data = iris)
abline(lm(Sepal.Length ~ Petal.Length, iris), col="blue")
```

```r
scatter.smooth(iris$Sepal.Length, iris$Petal.Length, family='gaussian', col='red')
```
